## Demo: How to work with database

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite.1") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = "SELECT COUNT(*)FROM master"

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
qry="Select ConvertedComp from master "
df=pd.read_sql_query(qry,conn)
df.plot(kind='hist')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
qry="Select Age from master "
df=pd.read_sql_query(qry,conn)
df.plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

qry="Select Age,WorkWeekHrs from master "
df=pd.read_sql_query(qry,conn)
df.plot.scatter(x='Age',y='WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
qry="Select Age,WorkWeekHrs,CodeRevHrs from master "
df=pd.read_sql_query(qry,conn)
#df.plot.scatter(x='WorkWeekHrs',y='CodeRevHrs',s='Age')
sn.scatterplot(data=df,x='WorkWeekHrs',y='CodeRevHrs',size='Age')
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes her
qry="select * from DatabaseDesireNextYear"
df=pd.read_sql_query(qry,conn)
top5db=df.groupby('DatabaseDesireNextYear').count().sort_values('Respondent',ascending=False).head(5)
top5db.plot.pie(y='Respondent', figsize=(10,6),legend=False,autopct='%0.0f%%')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
qry="select WorkWeekHrs,CodeRevHrs,Age from master"
df=pd.read_sql_query(qry,conn)
df=df[(df['Age']>=30)&(df['Age']<=35)]
df=df.groupby('Age').median()
df.plot.bar(stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
qry="select ConvertedComp,Age from master where Age>=45 and Age<=60"
df=pd.read_sql_query(qry,conn)
df=df.groupby('Age').median()
df.plot.line()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
qry="select MainBranch,Respondent from master"
df=pd.read_sql_query(qry,conn)
df=df.groupby('MainBranch').count()
df.plot.barh()
```
